```python
Data["timestamp"] = pd.to_datetime(Data["timestamp"],format='%Y年%m月%d日 %H:%M')
# 結果を表示
Data.head()
```

```python
Data.isnull().sum().sum()
```

```python
Data = Data.rename(columns={"Aouther": "Author"})
Data
```

```python
# Author列をカンマで分割して新しい列として追加
authors_split = Data['Author'].str.split(', ', expand=True)

# 列名を'Author1', 'Author2'のように設定
authors_split.columns = [f'Author{i+1}' for i in range(authors_split.shape[1])]

# 元のデータフレームと結合
Data = pd.concat([Data, authors_split], axis=1)
Data
```

```python
Data = Data.fillna('absent')
```

```python
Data['author_counts'] = (Data.loc[:, 'Author1':'Author11'] != 'absent').sum(axis=1)
```
