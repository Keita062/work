```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.preprocessing import StandardScaler
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(r"C:\Users\sk062\OneDrive\デスクトップ\Analitics\論文データ\Data_note.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"], format='%Y-%m-%d %H:%M:%S')
```

```python
scaler = StandardScaler()
df[['char_count_z', 'conclusion_char_count_z', 'author_counts_z']] = scaler.fit_transform(df[['char_count', 'conclusion_char_count', 'author_counts']])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))  # 図のサイズを指定
X = df["conclusion_char_count"]
Y = df["char_count"]
X2 = df["conclusion_char_count_z"]
Y2 = df["char_count_z"]

ax.scatter(X, Y, c="red", alpha=0.7, label="Character Count")  # alphaで視認性向上
ax.scatter(X2, Y2, c="blue", alpha=0.7, label="Character Count")  # alphaで視認性向上
ax.set_xlabel("Conclusion Character Count", fontsize=12)
ax.set_ylabel("Total Character Count", fontsize=12)

ax.grid(True, linestyle="--", alpha=0.6)  # グリッドを破線にして調整
ax.legend(loc="upper left")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))  # 図のサイズを指定

X2 = df["conclusion_char_count_z"]
Y2 = df["char_count_z"]

ax.scatter(X2, Y2, c="blue", alpha=0.7, label="Character Count")  # alphaで視認性向上
ax.set_xlabel("Conclusion Character Count", fontsize=12)
ax.set_ylabel("Total Character Count", fontsize=12)

ax.grid(True, linestyle="--", alpha=0.6)  # グリッドを破線にして調整
ax.legend(loc="upper left")

plt.show()
```

```python
# A列の0未満の数をカウント
count_negative_a = df['conclusion_char_count_z'][df['conclusion_char_count_z'] < 0].count()

# A列の0より高い数をカウント
count_positive_a = df['conclusion_char_count_z'][df['conclusion_char_count_z'] > 0].count()

print(f"0未満の数: {count_negative_a}")
print(f"0より高い数: {count_positive_a}")
```

```python
# A列の0未満の数をカウント
count_negative_b= df['char_count_z'][df['char_count_z'] < 0].count()

# A列の0より高い数をカウント
count_positive_b = df['char_count_z'][df['char_count_z'] > 0].count()

print(f"0未満の数: {count_negative_b}")
print(f"0より高い数: {count_positive_b}")
```

```python
print(df["char_count_z"].corr(df["conclusion_char_count_z"]))
print(df["conclusion_char_count_z"].corr(df["char_count_z"]))
```

```python
coef,p = stats.pearsonr(df["char_count_z"],df["conclusion_char_count_z"])
print("相関係数:",coef,"p-valu",p)
```
